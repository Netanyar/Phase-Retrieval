```python
# import useful modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
```

```python
# variables
imagedim = (100,100)
numdecimals = 5
filename = "starrynight"
```

```python
# Read image and create another to match provided requirements
img = Image.open(filename + ".jpg").resize(imagedim).convert('L')
img.save(filename+"edited.png")
```

```python
# load image into matplotlib and create matrix of numbers (scale values by 255 since they're initially in [0,1])
img = mpimg.imread(filename+"edited.png")*255
print("Current image:")
plt.imshow(img, cmap=plt.get_cmap('gray'))
```

```python
print("Matrix of values:")
print(img)
```

```python
# Computer fourier transform/absolute value:
b = np.abs(np.fft.fft2(img))
print(b)
```

```python
img_mat = np.mat(b)

# Write problem to text file
with open('problem.txt','wb+') as output:
    for line in img_mat:
        np.savetxt(output, line, fmt='%.'+str(numdecimals)+"f")
```
